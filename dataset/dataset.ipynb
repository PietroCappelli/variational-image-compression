```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def visualize_images(images):
    """Visualize images in a grid of 4 rows x 4 columns."""
    # Create a list of 16 images by random sampling from the dataset first dimension
    image_list = [images[i] for i in np.random.randint(0, len(images), 16)]

    # Create a 4x4 grid of images
    fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10, 10))
    for i, ax in enumerate(axes.flat):
        ax.imshow(image_list[i])
        ax.axis('off')

    # Show the plot
    plt.show()
```

```python
images = np.fromfile('../data/stl10/stl10_binary/unlabeled_X.bin', dtype=np.uint8)

images = images.reshape(-1, 3, 96, 96).transpose(0, 3, 2, 1) 

images.shape
```

```python
visualize_images(images)
```
